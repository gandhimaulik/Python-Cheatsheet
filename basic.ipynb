```python
my_list = ['A string',23,100.232,'o']
```

```python
my_list[1:]
```

**Dictonary**
* key value pair
* Order not maintained

```python
my_dict = {'key1':'value1','key2':'value2'}
my_dict.keys()
```

```python
my_dict.values()
```

```python
my_dict.items()
```

**Tuple**
* can't be changed

```python
t = (1,2,3)
```

**Sets**
* can contain only unique values
* it is like dictionary but only with keys

```python
list1 = [1,1,2,2,3,4,5,6,1,1]
set(list1)
```

>  **Statments**

```python
loc = 'Bank'

if loc == 'Auto Shop':
    print('Welcome to the Auto Shop!')
elif loc == 'Bank':
    print('Welcome to the bank!')
else:
    print('Where are you?')
```

```python
for num in list1:
    print(num)
```

```python
for letter in 'This is a string.':
    print(letter)
```

```python
list2 = [(2,4),(6,8),(10,12)]
#without unpacking
for tup in list2:
    print(tup)

#with unpacking
for (t1,t2) in list2:
    print(t1)    
```

```python
d = {'k1':1,'k2':2,'k3':3}
# Dictionary unpacking
for k,v in d.items():
    print(k)
    print(v)
```
